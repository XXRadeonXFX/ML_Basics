```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset= pd.read_csv('tips.csv')
```

```python
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset['tip'].fillna(dataset['tip'].mean(),inplace=True)
dataset['size'].fillna(dataset['size'].mean(), inplace=True)
```

```python
dataset.isnull().sum()
```

```python
dataset.info()
```

```python
# Corelation
data_corr= dataset.select_dtypes(['float64','int64']).corr()
```

```python
# Covariance
data_cov = dataset.select_dtypes(['float64','int64']).cov()
```

```python
plt.figure(figsize=(4,3))
sns.heatmap(data_corr ,annot=True)
plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.heatmap(data_cov, annot=True)

plt.show()
```
