```python
dataset = pd.read_csv('titanic.csv')
```

```python
dataset.head(3)
```

```python
dataset["Age"].mean()
```

```python
np.mean(dataset["Age"])
```

```python
sns.histplot(x="Fare" , data=dataset , bins=[ x for x in range(0,300,10) ])
plt.plot([ mn for i in range(0,300) ] , [ i for i in range(0,300) ] , c='yellow' , label = "Mean")
plt.plot([md for i in range(0,300) ] , [ i for i in range(0,300)] , label = "Median")
plt.plot([mo for i in range(0,300)], [i for i in range(0,300)], c='red', label = "Mode")
plt.legend()
plt.show()
```

```python
mn = np.mean(dataset["Fare"])
md = np.median(dataset["Fare"])
mo = dataset["Fare"].mode()

mn,md,mo
```
